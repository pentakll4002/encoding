# Nominal/OHE Encoding

##### Nominal encoding, also known as one-hot encoding, is a technique used to represent categorical data as numerical data, which is more suitable for machine learning algorithms. In this technique, each category is represented as a binary vector where each bit corresponds to a unique category. For example, if we have a categorical variable "color" with three possible values (red, green, blue), we can represent them as follows:

####Red: [1, 0, 0]
#### Green: [0, 1, 0]
####Blue: [0, 0, 1]

#### 3.Mã hóa Nominal/OHE

#### Mã hóa một không (nominal encoding), còn được gọi là mã hóa one-hot, là một kỹ thuật dùng để đại diện cho dữ liệu phân loại dưới dạng dữ liệu số, phù hợp hơn cho các thuật toán học máy. Trong kỹ thuật này, mỗi loại được đại diện bằng một vector nhị phân, nơi mỗi bit tương ứng với một loại duy nhất. Ví dụ, nếu chúng ta có một biến phân loại "màu sắc" với ba giá trị (đỏ, xanh lục, và xanh dương), chúng ta có thể biểu diễn như sau:

#### Đỏ: [1, 0, 0]
#### Xanh lục: [0, 1, 0]
#### Xanh dương: [0, 0, 1]

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
### Create a simple DataFrame
df = pd.DataFrame({
    'color': ['red', 'blue', 'green', 'green', 'red', 'blue']
})
```

```python
df.head()
```

```python
### Create an instance of OneHotEncoder
encoder = OneHotEncoder()
```

```python
### Perform fix and transform
### Bởi vì sắp xếp theo bảng chữ cái nên là mã hóa nó sẽ đi theo chữ cái luôn
encoded = encoder.fit_transform(df[['color']]).toarray()
```

```python
import pandas as pd
encoder_df = pd.DataFrame(encoded, columns=encoder.get_feature_names_out())
```

```python
encoder_df
```

```python
### For new data
encoder.transform([['blue']]).toarray()
```

```python
# Trong pd.concat([df, encoder_df], axis=1), tham số axis=1 có nghĩa là bạn đang nối các DataFrame theo cột.
pd.concat([df, encoder_df], axis=1)
```
